```python
# Option 1: Drop rows with missing values
df = df.dropna()

# OR Option 2: Fill missing values
# df['Age'] = df['Age'].fillna(df['Age'].median())
# df['Transaction_Amount'] = df['Transaction_Amount'].fillna(df['Transaction_Amount'].mean())
```

```python
df = df.drop_duplicates()
```

```python
sns.boxplot(df['Age'])
plt.show()

# Remove age outliers
df = df[df['Age'] <= 100]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['Spending_Score'] = scaler.fit_transform(df[['Spending_Score']])
```

```python
sns.histplot(df['Annual_Income'], bins=30, kde=True)
plt.title("Income Distribution")
plt.show()

sns.scatterplot(x='Annual_Income', y='Spending_Score', data=df)
plt.title("Income vs Spending")
plt.show()
```
